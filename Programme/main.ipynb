```python
import biblio_modules
import pyarrow
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
customers = pd.read_csv("../Source/customers.csv", sep=";")
products = pd.read_csv("../Source/products.csv", sep=";")
transactions = pd.read_csv("../Source/Transactions.csv", sep=";")

def load_csv_files_as_df(df, path, delim):
    df = pd.read_csv(path, sep=delim)
    return df
```

```python
transaction_product = pd.merge(transactions, products, how='outer', on='id_prod')
table_merged = pd.merge(transaction_product, customers, how='left', on='client_id')
table_merged
transaction_product
```
